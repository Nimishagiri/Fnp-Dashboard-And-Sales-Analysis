import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/fnp merge data.csv')
```

```python
data.describe()
```

described this data to understand order and product count
and to understand statistics

```python
data["Order_Date"] = pd.to_datetime(data["Order_Date"], format="%m/%d/%Y")
data["Delivery_Date"] = pd.to_datetime(data["Delivery_Date"], format="%m/%d/%Y")
```

```python
data["Delivery_diff"] = (data["Delivery_Date"] - data["Order_Date"]).dt.days
```

we achieved delivery difference of each order

```python
data.head()
```

**1.What is the total revenue generated?**

```python
revenue=data['Revenue'].sum()
print(f"Total Revenue Generated is {revenue}")
```

**2. Which product has the highest sales revenue?**

```python
product_revenue = data.groupby("Product_Name")["Revenue"].sum()
top_product = product_revenue.idxmax()
top_product_revenue = product_revenue.max()
print(f"Highest revenue product: {top_product} with â‚¹{top_product_revenue}")
```

```python
plt.figure(figsize=(10,5))
top_10_products = product_revenue.sort_values(ascending=False).head(10)
sns.barplot(y=top_10_products.index, x=top_10_products.values, palette="coolwarm")
plt.title("Top 10 Products by Revenue")
plt.xlabel("Total Revenue (INR)")
plt.ylabel("Product Name")
plt.show()
```

**3. Which city has the most orders?**

```python
city_orders = data["City"].value_counts()
top_city = city_orders.idxmax()
top_city_orders = city_orders.max()

print(f"City with the most orders: {top_city} with {top_city_orders} orders")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y=city_orders.head(5).index, x=city_orders.head(5).values, palette="viridis")
plt.title("Top 5 Cities by Number of Orders")
plt.xlabel("Number of Orders")
plt.ylabel("City")
plt.show()
```

**4. which category generates the most revenue?**

```python
category_revenue = data.groupby("Category")["Revenue"].sum()
top_category = category_revenue.idxmax()
top_category_revenue = category_revenue.max()

print(f"Category with the highest revenue: {top_category} with â‚¹{top_category_revenue}")
```

```python
plt.figure(figsize=(7,5))
top_categories = category_revenue.sort_values(ascending=False).head(10)
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%',
        colors=sns.color_palette("coolwarm"), startangle=140)
plt.title("Top Categories by Revenue")
plt.show()
```

**5.What is the average order value (AOV)?**

```python
revenue = data["Revenue"].sum()
orders = data["Order_ID"].nunique()
average_order_value = revenue / orders
print("Total Revenue: â‚¹", revenue)
print("Total Orders:", orders)
print("Average Order Value (AOV): â‚¹", round(average_order_value, 2))
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data["Revenue"], bins=30, kde=True, color="purple")
plt.title("Distribution of Order Value")
plt.xlabel("Revenue (INR)")
plt.ylabel("Frequency")
plt.show()
```

**6. What is the most ocassion for orders?**

```python
occasion_counts = data["Occasion"].value_counts()

top_occasion = occasion_counts.idxmax()
top_occasion_count = occasion_counts.max()

print(f"Most common occasion: {top_occasion} with {top_occasion_count} orders")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=occasion_counts.head().index, y=occasion_counts.head().values, palette="Set2")
plt.title("Top Occasions for Orders")
plt.ylabel("Number of Orders")
plt.xlabel("Occasion")
plt.xticks(rotation=45)
plt.show()
```

**7. which product is the ordered the most(by quantity)?**

```python
product_quantity = data.groupby("Product_Name")["Quantity"].sum()
top_product = product_quantity.idxmax()
top_product_quantity = product_quantity.max()

print(f"Most ordered product: {top_product} with {top_product_quantity} units sold")
```

```python
plt.figure(figsize=(12,6))
df_grouped = data.groupby("Order_Date")["Quantity"].sum().reset_index()
sns.lineplot(x="Order_Date", y="Quantity", data=df_grouped, marker="o", color="blue")
plt.title("Total Products Ordered Over Time")
plt.xlabel("Order Date")
plt.ylabel("Total Quantity Ordered")
plt.xticks(rotation=45)
plt.show()
```

**8.Customer placed the most orders?**

```python
customer_orders = data["Customer_ID"].value_counts()
top_customer_id = customer_orders.idxmax()
top_customer_orders = customer_orders.max()


top_customer_name = data.loc[data["Customer_ID"] == top_customer_id, "Name"].iloc[0]

print(f"Customer with the most orders: {top_customer_name} ({top_customer_id}) with {top_customer_orders} orders")
```

```python
plt.figure(figsize=(8,8))
top_5_customers = customer_orders.head(5)
plt.pie(top_5_customers, labels=top_5_customers.index, autopct='%1.1f%%',
        colors=sns.color_palette("pastel"), startangle=140)
plt.title("Top 5 Customers by Orders")
plt.show()
```

**9.Which is the average delivery time?**

```python
average_delivery_time = data["Delivery_diff"].mean()

print(f"Average Delivery Time: {round(average_delivery_time, 2)} days")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=data["Delivery_diff"], color="orange")
plt.title("Boxplot of Delivery Time")
plt.xlabel("Delivery Days")
plt.show()
```

**10. Which month had highest number of orders?**

```python
data["Order_Month"] = data["Order_Date"].dt.month
```

```python
data.head(5)
```

```python
monthly_orders = data["Order_Month"].value_counts().sort_index()
top_month = monthly_orders.idxmax()
top_month_orders = monthly_orders.max()

print(f"Month with the highest number of orders: {top_month} with {top_month_orders} orders")
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_orders.index, y=monthly_orders.values, marker="o", color="purple")
plt.title("Monthly Order Trend")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(True)
plt.show()
```
